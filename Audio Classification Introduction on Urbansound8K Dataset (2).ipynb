```python
                                                   # Audio Classification Introduction
import librosa
import numpy as np
import matplotlib.pyplot as plt
#Displays static plots in the notebook.
%matplotlib inline 
#    %matplotlib qt- to Open plots in an external window.
import IPython.display as ipd
import librosa.display
from scipy.io import wavfile as wav
```

```python
file_path="C:/Users/mukhe/Downloads/UrbanSound8K/UrbanSound8K/audio/fold1/98223-7-3-0.wav"
```

```python
plt.figure(figsize=(14,5))
data,sample_rate=librosa.load(file_path)
librosa.display.waveshow(data,sr=sample_rate)
ipd.Audio(file_path)
```
